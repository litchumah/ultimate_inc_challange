## Part 1 ‑ Exploratory data analysis

The attached logins.json file contains (simulated) timestamps of user logins in a particular geographic location.
```python
df = pd.read_json('logins.json')
df.head()
```

By aggregating the login times in 15-minute time intervals it possible to observe that there is an increase in the number of logins throughout time. Also, by observing daily login count, it possible to notice that the peaks are in the very early morning and late night, between 23:00 and 5:00

```python
df2 = pd.read_json('ultimate_data_challenge.json')
df2.head()
```

```python
df2.info()
```

```python
df2 = df2.drop(df2[df2.phone.isna()==True].index,axis=0)
df2.info()
```

```python
df2['avg_rating_of_driver'].describe()
```

```python
df2['avg_rating_of_driver']= df2['avg_rating_of_driver'].fillna(value=df2['avg_rating_of_driver'].mean())
```

```python
df2.info()
```

```python
df2['avg_rating_by_driver']= df2['avg_rating_by_driver'].fillna(value=df2['avg_rating_by_driver'].mean())
```

```python
df2.info()
```

```python
df2.head()
```

```python
df2['last_trip_date'] = pd.to_datetime(df2['last_trip_date'])
current_date = df2['last_trip_date'].max()
df2['active'] = df2['last_trip_date'].apply(lambda date: 0 if (current_date-date).days>30 else 1)
df2.head()
```

```python
#create dummy variables for categorical features
df_features = pd.get_dummies(df2,columns=['city','phone','ultimate_black_user'])
#set target and drop unuseful columns
target = df_features['active']
df_features = df_features.drop(['last_trip_date','signup_date','active'],axis=1)
```

```python
df_features.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import QuantileTransformer, StandardScaler
from pipelinehelper import PipelineHelper
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_features, target, test_size=0.3,random_state=42)
```

```python
pipe = Pipeline([
    ('scaler', PipelineHelper([
        ('std', StandardScaler()),
        ('qnt', QuantileTransformer()),
    ])),
    ('classifier', PipelineHelper([
        ('lr', LogisticRegression()),
        ('rf', RandomForestClassifier()),
    ])),
])

params = {
    'scaler__selected_model': pipe.named_steps['scaler'].generate({
        'std__with_mean': [True, False],
        'std__with_std': [True, False],
    }),
    'classifier__selected_model': pipe.named_steps['classifier'].generate({
        'lr__C': [0.1, 1.0],
        'lr__solver':['liblinear'],
        'rf__n_estimators': [100, 20],
    })
}
grid = GridSearchCV(pipe, params, scoring='accuracy', verbose=1)
grid.fit(X_train, y_train)
```

```python
grid.score(X_test, y_test)
```

```python
print(cross_val_score(grid,X_test,np.ravel(y_test),cv=5))
```

```python
print (grid.best_params_)
```
